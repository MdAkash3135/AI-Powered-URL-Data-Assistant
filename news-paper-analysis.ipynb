```python
from langchain_community.document_loaders import UnstructuredURLLoader
```

```python
loader = UnstructuredURLLoader(
    urls=['https://en.prothomalo.com/sports/football/4uidp3kbbw', 'https://en.prothomalo.com/sports/cricket/pa75n0nzp4' ]

)
```

```python
data = loader.load()
data[0].page_content[:500]
```

```python
text = data[0].page_content
print(text[:1000])
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter_using_charTextSplitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 200,
    chunk_overlap  = 40,
    length_function = len,
)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter_using_RecursiveCharacterTextSplitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n", "\n", " ", ""],
    chunk_size = 250,
    chunk_overlap = 20,
    length_function = len,
)
```

```python
chunks = text_splitter_using_RecursiveCharacterTextSplitter.split_text(text)
print(f"Number of chunks: {len(chunks)}")
for i, chunk in enumerate(chunks):
    print(f"--- Chunk {i+1} ---")
    print(len(chunk))
    
```

```python
from sentence_transformers import SentenceTransformer

encoder = SentenceTransformer('all-MiniLM-L6-v2')

vectors = encoder.encode(chunks)
vectors.shape
```

```python
import faiss
index = faiss.IndexFlatL2(vectors.shape[1])


```

```python
index.add(vectors)
```

```python
query_search = "Defeats by Chelsea and Paris Saint-Germain left Barca in a difficult position in Europe, but by beating Frankfurt they are still hopeful of avoiding the play-off round." 
```

```python
svec = encoder.encode([query_search])
svec.shape
```

```python
index_list = index.search(svec, k = 1)
print(index_list[1][0])
```
